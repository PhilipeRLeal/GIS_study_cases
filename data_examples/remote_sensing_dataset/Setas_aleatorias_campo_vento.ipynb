```python
import numpy as np

import matplotlib.pyplot as plt

import cartopy.crs as ccrs
```

```python
A = np.random.randn(11, 11)

X = np.linspace(0, 10, 11)
Y = np.linspace(0, 10, 11)

x2d, y2d = np.meshgrid(X, Y)


u, v = np.gradient(A)


fig = plt.figure(num=1, figsize=(15, 15))
ax1 = fig.add_subplot(3, 3, 1, projection=ccrs.PlateCarree())
ax1.set_title("Vector Field")
ax1.set_extent([-1, 11, -1, 11], ccrs.PlateCarree())
ax1.set_xticks(ticks=list(np.arange(11)))
ax1.set_yticks(ticks=list(np.arange(11)))
ax1.quiver(x2d, y2d, u, v)
ax1.imshow(A)
plt.colorbar(ax1.imshow(A))
plt.show()
```
