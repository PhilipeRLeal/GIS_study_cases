# para determinar a coordenada de qualquer píxel em um arquivo raster, deve-se utilizar de uma transformação linear da seguinte maneira:

Latitude = Linha*resolução Y +latitude inicial + (resolução Y * 0.5)

Longitude = coluna * resolução X + longitude inicial + (resolução X *0.5)

Notar que foi adiionado (resolução*0.5) aos valores estimados, pois a GDAL indica o canto do píxel e desta maneira podemos indicar a coordenada do centro de um píxel
