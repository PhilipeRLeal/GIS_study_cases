```python
# -*- coding: utf-8 -*-
"""
Created on Thu May 04 14:49:28 2017

@author: Philipe Leal
"""

#importacao de bibliotecas
import numpy as np
from osgeo import gdal
from osgeo import osr

# Definicao dos valores máximos médios e mínimos anuais de toda a quadrícula.


n = np.arange(1,13)

NS=[]

for x in n:
    NS.append(str(x))
    
    
Lista2 = []

for x in NS[0:9]:
    Lista2.append("0" + x)

    
print(Lista2)

NS[0:9] = Lista2

print("NS = {0}".format(NS))

focos = []

for Numero in NS:
       Nomedoarquivo = "2010_" + Numero + "_grd_focos.bin"
       focos.append(np.fromfile(r"C:/Users/Philipe Leal/Python/Exercicios_Logica_Programacao/focos/" + Nomedoarquivo, dtype=np.int16))

print(focos)
arr1 = np.array(focos)
arr2 = arr1.reshape(12,6300, 5000)
 
Max_anual_de_toda_matriz = np.max(focos)

Min_anual_de_toda_matriz = np.min(focos)

Mean_anual_de_toda_matriz = np.mean(focos)


print("maximo_anual_de_toda_matriz = {0}".format(Max_anual_de_toda_matriz))
print("")
print("minimo_anual_de_toda_matriz = {0}".format(Min_anual_de_toda_matriz))
print("")
print("media_anual_de_toda_matriz = {0}".format(Mean_anual_de_toda_matriz))

# Definindo parametros da projecao e da geotransformacao a serem utilizados na construcao dos arquivos matriciais:

geotransform = (-83.005, 0.01, 0.0, 13.005, 0.0, -0.01)

#geotransform[0] /* top left x 
#geotransform[1] /* w-e pixel resolution 
#geotransform[2] /* 0 
#geotransform[3] /* top left y 
#geotransform[4] /* 0 
#geotransform[5] /* n-s pixel resolution (negative value) 

projecao = osr.SRS_WKT_WGS84

# definicao dos parametros do arquivo de saida em GTIFF
#notar que o tamanho do arquivo de saida do GTIFF é maior que o somatorios dos aquivos de ENVI somados entre si. Portanto, o metadados do gtiff é mais descritivo do que o do envi.

driver = gdal.GetDriverByName ("GTiff") # nome do formato dos arquivos de saida
xsize = 5000
ysize =6300
nbandas = 1

# nomes_arquivo_saida = "C:/Users/Philipe Leal/Python/exercicios/focos/"  OU USAR COMANDO ABAIXO:
nome_arquivo_saida_max = r"C:/Users/Philipe Leal/Python/Exercicios_Logica_Programacao/focos/Anual_Max.tif"
nome_arquivo_saida_min = r"C:/Users/Philipe Leal/Python/Exercicios_Logica_Programacao/focos/Anual_Min.tif"
nome_arquivo_saida_mean = r"C:/Users/Philipe Leal/Python/Exercicios_Logica_Programacao/focos/Anual_Mean.tif"

# definicoes dos arquivos de saida
File_output_Max = driver.Create(nome_arquivo_saida_max, xsize, ysize, nbandas, gdal.GDT_Int16)
File_output_Min = driver.Create(nome_arquivo_saida_min, xsize, ysize, nbandas, gdal.GDT_Int16)
File_output_Mean = driver.Create(nome_arquivo_saida_mean, xsize, ysize, nbandas, gdal.GDT_Int16)

#definindo set de projecoes e transformacoes nos arquivos de saida

File_output_Max.SetGeoTransform(geotransform)
File_output_Max.SetProjection(projecao)

File_output_Min.SetGeoTransform(geotransform)
File_output_Min.SetProjection(projecao)

File_output_Mean.SetGeoTransform(geotransform)
File_output_Mean.SetProjection(projecao)

#Definindo valores de max min e mean anuais para cada pixel da matriz:


Max_anual_por_pixel = arr2.max(axis = 0)

Min_anual_por_pixel = arr2.min(axis = 0)

Mean_anual_por_pixel = arr2.mean(axis = 0)
    
#atribuindo valores ao Gtiff:
    
File_output_Max.GetRasterBand(1).WriteArray(Max_anual_por_pixel)
File_output_Max = None

File_output_Min.GetRasterBand(1).WriteArray(Min_anual_por_pixel)
File_output_Min = None

File_output_Mean.GetRasterBand(1).WriteArray(Mean_anual_por_pixel)
File_output_Mean = None
```
